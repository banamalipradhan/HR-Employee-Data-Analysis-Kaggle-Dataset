**Q.15) Identify departments with the highest attrition rate (Resigned %).**

```python
df.head(2)
```

```python
#displays total employees and resigned employees on each department 

dept_counts=df.groupby("Department")["Status"].agg(Total_Employees="count",Resigned_Employees=lambda x: (x=="Resigned").sum())
dept_counts
```

```python
#creating a new column attrition_rate which displays the % of resigned employees 

dept_counts["Attrition_Rate%"] = (dept_counts["Resigned_Employees"]/dept_counts["Total_Employees"])*100
dept_counts
```

```python
#sorting the attrition_rate column in descending order 

dept_counts.sort_values("Attrition_Rate%",ascending=False)
```

```python
plt.figure(figsize=(12, 6))

# X positions
x = np.arange(len(dept_counts))
width = 0.35

# Bar plots (Total & Resigned)
plt.bar(x - width/2, dept_counts["Total_Employees"], width, label="Total Employees")
plt.bar(x + width/2, dept_counts["Resigned_Employees"], width, label="Resigned Employees")

# Line plot (Attrition Rate %)
plt.plot(
    x,
    dept_counts["Attrition_Rate%"],
    color="black",
    marker="o",
    linewidth=2,
    label="Attrition Rate (%)"
)

# X-axis labels from index (Department)
plt.xticks(x, dept_counts.index, rotation=45)

# Labels and title
plt.ylabel("Employees / Attrition Rate")
plt.title("Employee Overview by Department")

plt.grid(axis="y", linestyle="--", alpha=0.4)
plt.legend()
plt.tight_layout()

plt.show()
```
